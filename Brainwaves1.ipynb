```
import numpy as np
import pandas as pd

x=pd.read_csv("train.csv")
y=pd.read_csv("test.csv")
```

```
z=x["Complaint-Status"]
x
```

```
x=x.drop(['Complaint-Status'],axis=1)
x=x.append(y,ignore_index=True)
```

```
x["Consumer-disputes"]=x["Consumer-disputes"].replace(["Yes","No"],[0,1])
x["Consumer-disputes"].fillna(0,inplace=True)
x
```

```
k=x['Transaction-Type'].unique()
```

```
x['Transaction-Type']=x['Transaction-Type'].replace(k,[i for i in range(len(set(x['Transaction-Type'])))])
```

```
k=x['Complaint-reason'].unique()
x['Complaint-reason']=x['Complaint-reason'].replace(k,[i for i in range(len(set(x['Complaint-reason'])))])
```

```
k=x['Company-response'].unique()
x['Company-response']=x['Company-response'].replace(k,[i for i in range(len(set(x['Company-response'])))])
```

```
x=x.drop(['Date-received'],axis=1)
x=x.drop(['Date-sent-to-company'],axis=1)
x=x.drop(['Consumer-complaint-summary'],axis=1)
```

```
x
```

```
k=z.unique()
z=z.replace(k,[i for i in range(len(set(z)))])
print(z)
```

```
from sklearn import *
from sklearn.model_selection import train_test_split
full_X=x.iloc[:,1:]
full_X=full_X.iloc[:,:]
train_valid_X = full_X[ 0:43266 ]
train_valid_y = z
test_X = full_X[ 43266: ]
train_X , valid_X , train_y , valid_y = train_test_split( train_valid_X , train_valid_y , train_size = .7 )

print (full_X.shape , train_X.shape , valid_X.shape , train_y.shape , valid_y.shape , test_X.shape)
```

```
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from xgboost import XGBClassifier
model = LogisticRegression()
```

```
train_X=train_X.astype(float)
train_y=train_y.astype(float)
model.fit(train_X,train_y)
```

```
valid_X=valid_X.astype(float)
valid_y=valid_y.astype(float)
print (model.score( train_X , train_y ) , model.score( valid_X , valid_y ))
```

```
#['Closed with explanation', 'Closed with non-monetary relief',
#       'Closed', 'Closed with monetary relief', 'Untimely response']
test_Y = model.predict(test_X).astype(int)
print(test_Y)
Complaint-ID = x[43266:].Complaint-ID
test = pd.DataFrame( { 'Complaint-ID': Complaint-ID , 'Complaint-Status': test_Y })
test.shape
test.head()
test.to_csv( 'test_csv.csv' , index = False )
```
